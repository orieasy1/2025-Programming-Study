#### Exercise 3. print 놀이

```python
print("3. print 놀이")

for i in range(5):
    for j in range(4-i):
        print(" ", end = "")
    for j in range(2*i+1):
        print("*", end = "")
    print();

```

#### Exercise 4. 데이터시각화 따라해보기

아래 두 사이트 중 한 곳을 들어가서(클릭!), 마음에 드는 예제를 복사해 붙여넣어 실행해보세요 
+ [Matplotlib](https://matplotlib.org/stable/gallery/index.html)
+ [Seaborn](https://seaborn.pydata.org/examples/index.html)

```python
print("4. 데이터시각화 따라해보기")
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(-10, 10, 100)
sig = 1 / (1 + np.exp(-t))

fig, ax = plt.subplots()
ax.axhline(y=0, color="black", linestyle="--")
ax.axhline(y=0.5, color="black", linestyle=":")
ax.axhline(y=1.0, color="black", linestyle="--")
ax.axvline(color="grey")
ax.axline((0, 0.5), slope=0.25, color="black", linestyle=(0, (5, 5)))
ax.plot(t, sig, linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")
ax.set(xlim=(-10, 10), xlabel="t")
ax.legend(fontsize=14)
plt.show()
```

#### Exercise 5. 데이터시각화 바꿔보기

```python
print("5. 데이터시각화 바꿔보기")
print("그래프의 색깔을 바꿨습니다.")

import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(-10, 10, 100)
sig = 1 / (1 + np.exp(-t))

fig, ax = plt.subplots()
ax.axhline(y=0, color="blue", linestyle="--")
ax.axhline(y=0.5, color="blue", linestyle=":")
ax.axhline(y=1.0, color="blue", linestyle="--")
ax.axvline(color="green")
ax.axline((0, 0.5), slope=0.25, color="red", linestyle=(0, (5, 5)))
ax.plot(t, sig, linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")
ax.set(xlim=(-10, 10), xlabel="t")
ax.legend(fontsize=14)
plt.show()
```
